#### H0 : Airline has no effect on price

```python
# Airline vs Price
groups = [group["price"].values for _, group in df.groupby("airline")]
res = stats.f_oneway(*groups)
p_value = res.pvalue
alpha = 0.05
print(f"p-value: {p_value}")
if p_value < alpha:
    print("Reject H0 → airline significantly affects price ")
else:
    print("Fail to reject H0 → no significant effect ")
```
