## TESTS    


#### H0 : Airline has no effect on price

```python
# Airline vs Price
groups = [group["price"].values for _, group in df.groupby("airline")]
res = stats.f_oneway(*groups)
p_value = res.pvalue

alpha = 0.05
print(f"p-value: {p_value}")
if p_value < alpha:
    print("Reject H0 → airline significantly affects price ")
else:
    print("Fail to reject H0 → no significant effect ")
```

# Anova test
source_city gouped with destination_city vs price


```python
#source_des =  df.groupby([df["source_city"], df["destination_city"]])

groups = [group["price"].values 
        for _, group in df.groupby(["source_city", "destination_city"])]
res = stats.f_oneway(*groups)
p_value = res.pvalue
alpha = 0.05
print(f"p-value: {p_value:.3f}")
if p_value < alpha:
    print("Reject H0 → route significantly affects price ")
else:
    print("Fail to reject H0 → no significant effect ")
```

```python
airlines = df["airline"].unique().tolist()
dict_airline_prices = {}
for airline in airlines :
    airline_prices = df[df['airline'] == airline]['price']
    dict_airline_prices[airline] = airline_prices

anova_results = sp.stats.f_oneway(*dict_airline_prices.values())
print(f"P-value:{anova_results[1]:.10f}")
```
