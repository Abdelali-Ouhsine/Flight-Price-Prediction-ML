## Machine Learning Pipeline :

### Phase 0 : Import libraries: 

```python

import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.gridspec import GridSpec
```

### Phase 1 : Import Data :


```python
df = pd.read_csv("Clean_Dataset.csv")
df.head()
```

### Phase 2 : EDA 

## T_test


```python
economy_prices = df[df["class"] == "Economy"]["price"]
business_prices = df[df["class"] == "Business"]["price"]
t_stat, p_value = stats.ttest_ind(economy_prices, business_prices, equal_var=False)
print(f"T-test: t-statistic = {t_stat:.2f}, p-value = {p_value:.4f}")
```

# TEST 2 — One-Way ANOVA: Airline vs Price

```python
airlines = df["airline"].unique()
groups = [df[df["airline"] == a]["price"].values for a in airlines]
f_stat, p_value2 = stats.f_oneway(*groups)
print(f"ANOVA test: F-statistic = {f_stat:.2f}, p-value = {p_value2:.4f}") 
```

## TEST 3 — Chi-Square Test: Stops vs Class

```python
contingency_table = pd.crosstab(df["stops"], df["class"])
chi2_stat, p_value3, dof, expected = stats.chi2_contingency(contingency_table)
print(p_value3)
```

##  TEST 4 — Pearson Correlation: Duration vs Price

```python
r, p_value4 = stats.pearsonr(df["duration"], df["price"])
print(f"Pearson correlation: r = {r:.4f}, p-value = {p_value4:.4f}")
```
